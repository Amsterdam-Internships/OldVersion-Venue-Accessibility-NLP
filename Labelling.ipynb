```python
def classify_accessibility(data):
    '''Classifies text as discussing accessibility very sure.'''
    
    accessibility_terms = ["wheelchair", #119
                           "disability", #3
                           "disabled", #56
                           "handicap", #19
                           
                           #"mobility",
                           
                           "narrow entrance", #2
                           "wide entrance", #1
                           
                           "narrow stair", #10
                           "dangerous stair", #2
                           "elevator", #36
                           "steep", #235
                           
                           "slippery", #14
                           "cramped", #293
                           "spacious",
                           
                           "accessible toilet", #3
                           "accessible bathroom", #2
                           "toilet is upstairs",
                           "toilet is downstairs",
                           "bathroom is upstairs"] #917

    if any(i in data for i in accessibility_terms):
        return 1
    else:
        return 0
```

```python
%%time
data["Accessibility"] = data["Text"].apply(classify_accessibility)

# How any are classified as containing accessibility information?
print("Accessibility == 1:", data[data["Accessibility"]==1].shape[0])
```
